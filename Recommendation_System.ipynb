```python
#importing pandas library for data handling and manipulation
import pandas as pd 
#importing numpy for scientific calculation 
import numpy as np
```

```python
book = pd.read_csv('/content/book (1).csv', encoding='latin-1')
book.head()
```

```python
book.sort_values('User.ID') #sorting data based on userId
```

```python
book1=book.rename({'User.ID':'UserID','Book.Rating':'BookRating','Book.Title':'BookTitle'},axis=1)
```

```python
#here we are creating pivot table 
user_book1_df = book1.pivot_table(index='UserID',
                                 columns='BookTitle',
                                 values='BookRating')
```

```python
user_book1_df
```

```python
#Impute those NaNs with 0 values
user_book1_df.fillna(0, inplace=True)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances #for calculating distance of similarity
from scipy.spatial.distance import cosine, correlation #for correlation between users
```

```python
user_sim = 1 - pairwise_distances( user_book1_df.values,metric='cosine') # highly related will be 1 and not highly related will be less than one 
#and we took default one because cos 90 is one
```

```python
user_sim # we got array of that same dataste
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Impute those NaNs with 0 values
user_book1_df.fillna(0, inplace=True)
```
